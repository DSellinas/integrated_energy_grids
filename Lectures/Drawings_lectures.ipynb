# Drawings from lectures

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = [3, 3]
plt.rcParams["figure.autolayout"] = True

def f(x):
   return (x-5)**2 

x = np.linspace(0, 10, 100)

plt.plot(x, f(x), color='blue')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0,10])
plt.ylim([0,25])
plt.savefig('fig1.png')
```

```python
plt.rcParams["figure.figsize"] = [3, 3]
plt.rcParams["figure.autolayout"] = True

def f(x):
   return x**4-(7*x)**2+2000 

x = np.linspace(-10, 10, 100)

plt.plot(x, f(x), color='blue')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([-10,10])
plt.ylim([0,6000])
plt.savefig('fig2.png')
```

```python
plt.rcParams["figure.figsize"] = [3, 3]
plt.rcParams["figure.autolayout"] = True

def f(x):
   return x**3

x = np.linspace(-10, 10, 100)

plt.plot(x, f(x), color='blue')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([-10,10])
plt.ylim([-1000,1000])
plt.savefig('fig3.png')
```

```python
plt.rcParams["figure.figsize"] = [3, 3]
plt.rcParams["figure.autolayout"] = True

def f(x):
   return (x-5)**2
x = np.linspace(0, 10, 100)
plt.plot(x, f(x), color='blue')


colors=['gold', 'darkorange', 'red']
for i,k in enumerate([2, 0.5, 0.1]):
    def f(x):
       return (x-5)**2 -k*np.log(x-6)
    plt.plot(x, f(x), color=colors[i], label=k)
    
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0,10])
plt.ylim([0,25])
plt.legend(title='k')
plt.savefig('fig4.png')
```

```python
plt.rcParams["figure.figsize"] = [3, 3]
plt.rcParams["figure.autolayout"] = True


colors=['gold', 'darkorange', 'red']
for i,k in enumerate([2, 0.5, 0.1]):
    def f(x):
       return -k*np.log(x-6)
    plt.plot(x, f(x), color=colors[i], label=k)
    
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0,10])
plt.ylim([0,5])
plt.legend(title='k')
plt.savefig('fig5.png')
```
